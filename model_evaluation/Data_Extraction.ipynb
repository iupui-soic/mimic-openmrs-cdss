#SQL Queries for data extraction
1.	For 10K Diagnosis:

SELECT person_id, concept_id, value_numeric, value_coded
FROM obs
JOIN (
  SELECT DISTINCT patient_id
  FROM encounter_diagnosis
  WHERE diagnosis_coded IN (1385187,1385109,1384870,1384841,1384995,1397715,1384886,1474651,1385083,1385121)
  ORDER BY patient_id DESC
  LIMIT 12080
) a
ON obs.person_id = a.patient_id
ORDER BY person_id;


2.	For 10k patients obs:

SELECT encounter_diagnosis.patient_id, encounter_diagnosis.diagnosis_coded 
FROM encounter_diagnosis 
JOIN (
  SELECT DISTINCT patient_id 
  FROM encounter_diagnosis 
  WHERE diagnosis_coded IN (1385187,1385109,1384870,1384841,1384995,1397715,1384886,1474651,1385083,1385121)
  ORDER BY patient_id 
) a 
ON encounter_diagnosis.patient_id = a.patient_id 
ORDER BY diagnosis_coded;
